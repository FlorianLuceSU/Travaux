On note $\|.\|$ la norme euclidienne.
Pour $A\in{\cal M}_{m,n}(\mathbb{R})$, on considère le problème de minimisation 
$$\begin{equation*}
\min_{x\in C} f(x), 
\end{equation*}$$
où 
$$\begin{equation*}
f: x\in C\subset{\mathbb R}^n \mapsto f(x)=\|Ax\|.
\end{equation*}$$

On se propose de résoudre le problème dans le cas où $C$ est la sphère unité de $\mathbb{R}^n$. 
Si la matrice $A$ n'est pas injective, le minimum est égal à $0$. On suppose par la suite que $A$ est injective. 

*Note : minimiser $f$ revient à minimiser $f^2$ qui est une fonction différentiable.* 


- Modifier la fonction `Gradient` en une fonction `GradientProjete` pour résoudre le problème ci-dessus par la méthode du gradient projeté.

```python
def gradient (GradF, v0, rho,tol, NitMax):
    v=v0.copy()
    gradfx=GradF(v)
    it=0
    X = [v0]
    while norm(gradfx) > tol and it <NitMax: #critere d’arret sur le gradien
        #print(v,gradfx)
        v=v-rho*gradfx
        X.append(v) # on sauve l’iteration
        gradfx=GradF(v)
        it+=1
    cvg = norm(gradfx) <= tol #on verifie si on a convergence
    return v,X,it,cvg
```

Fonctions

```python
def facto(n):
    """
    calcul de la factorielle de l'entier n
    """ 
    if type(n)!= int :
        print(f"Ce n'est pas un nombre entier : {n}")
    else:
        if n < 0:
            print(f"entier negatif : {n}")
        else:
            c = 1
            for i in range(2, n+1):
                c = c * i
            return c
```

### EDO scalaire du premier ordre

### EDO vectorielle du premier ordre
Résoudre le système d'EDO (équations de Lorenz).
\[\left\{\begin{array}{rcl}
x'(t)&=&\sigma(y(t)-x(t)), 
 
y'(t)&=&-x(t)z(t)+\varrho x(t)-y(t)), 
 
z'(t)&=&x(t)y(t)-\beta z(t))
.\end{array}\right..\]

```python
from scipy.integrate import odeint

def f(u, t):  # meme si la fonction ne depend pas de t
    x, y, z = u
    f1 = sigma * (y-x)
    f2=-x * z+r * x-y
    f3 = x * y-b * z
    return (f1, f2, f3)
#def f(u, t, arg=(sigma, r, b)):    # autre possibilite
```

```python
sigma, r, b = 10, 28, 8/3
u0=[0, 1, 1.05]
Tf= 20   #temps final 
Temps = np.linspace(0, Tf, 100)
sol = odeint(f, u0, Temps) 
plt.plot(Temps, sol[:, 0], '-r', Temps, sol[:, 1], 'b*', Temps, sol[:, 2], 'y+')
plt.legend(['x', 'y'])
plt.show()
```

### EDO scalaire du second ordre
On se ramène à une EDO vectorielle du premier ordre. 

Exemple :
$x''(t)+x'(t)+x(t)=1$ et les conditions initiales $x(0)=1$, $x'(0)=2$ qui devient 
$u''(t)+Au(t)+b = 0$ avec $u(t)=(x(t), x'(t))^T$, 
$A=\begin{pmatrix} 0 & -1 
 1 & 1 \end{pmatrix}$ et $b=(0, 1)^T$ et la condition initiale 
$u(0)=(1, 2)^T$.

```python
def f2(u, t):  # meme si la fonction ne depend pas de t
    x, y = u
    f1 = y
    f2 = 1-x-y
    return ([f1, f2])
```

```python
u0=[1, 2] #donnee initiale
Tf= 10   #temps final
Temps = np.linspace(0, Tf, 100)
sol = odeint(f2, u0, Temps)
plt.plot(Temps, sol[:, 0], '-r', Temps, sol[:, 1], 'b*')
plt.legend(["x(t)", "x'(t)"])
plt.show()
```

## Pour aller plus loin: introduction aux réseaux de neurones
Un réseau de neurones est une fonction paramétrique qui s'écrit comme une succession d'opérations linéaires et non-linéaires.
Par exemple, la fonction $f(x) = w \cdot
\sigma(Wx + b)$ où $x \in \mathbb{R}^n$ est l'entrée, $w \in \mathbb{R}^h$, $W \in \mathbb{R}^{h \times n}$ et $b \in \mathbb{R}^h$ sont des paramètres, $\sigma$ est une fonction non-linéaire et $\cdot$ est le produit scalaire.

Dans cet exemple $\sigma$ peut être la fonction sigmoide $\sigma(x) = \frac{1}{1 + e^{-x}}$ ou bien la fonction $ReLU$ $\sigma(x) = \max(0, x)$.

Pytorch nous permet de definir facilement cette fonction etde calculer automatiquement le gradient d'une fonction paramétrique $f$ par rapport aux paramètres $w$, $W$ et $b$.
